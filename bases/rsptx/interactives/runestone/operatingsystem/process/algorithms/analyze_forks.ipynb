```python
with open('fork_variability_results.json', 'r') as f:
    data = json.load(f)

individual_results = data['individualResults']
```

```python
df_stats = pd.DataFrame({
        "code": [],
        "beforeWaitPrints": [],
        "afterWaitPrints": [],
        "childPrints": [],
        "otherPrints": [],
        "maxDepth": [],
        "totalForks": []
})

for _, value in individual_results.items():
    for item in value:
        df_stats.loc[len(df_stats)] = [item['code'], item['stats']['beforeWaitPrints'], item['stats']['afterWaitPrints'], item['stats']['childPrints'], item['stats']['otherPrints'], item['stats']['maxDepth'], item['stats']['totalForks']]

df_stats.head(5)
```

```python
# Plot histograms of the stats for totalForks == 2

df_stats_fork2 = df_stats[df_stats['totalForks'] == 2]

# Print the summary statistics
print(df_stats_fork2.describe())

# Plot histograms of the stats for totalForks == 2

# Create a figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Distribution of Process Statistics (Percentages)', fontsize=16)

# Flatten axes for easier iteration
axes = axes.flatten()

# Plot histograms for each column
columns = ['beforeWaitPrints', 'afterWaitPrints', 'childPrints', 'otherPrints', 'maxDepth']
for idx, col in enumerate(columns):
    # Calculate weights to convert to percentages
    weights = [100.0 / len(df_stats_fork2)] * len(df_stats_fork2)
    df_stats_fork2[col].hist(ax=axes[idx], bins=20, weights=weights)
    axes[idx].set_title(f'{col} Distribution')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Percentage (%)')

# Remove the empty subplot
axes[-1].remove()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
